# 3. Negativity tests

Remember that the negativity of a normal (diagonalizable) operator is defined as follows:

>\begin{equation}
    n(\rho) = \sum _{i:\lambda _i <0} |\lambda _i|
\end{equation}

Where $\lambda _i$ are the eigenvalues of the partial transpose of $\rho$.

## 3.3. Some observations

### 3.3.1.

Recordemos que las matrices densidad son operadores positivos. Por tanto, han de tener negatividad cero. Sin embargo, si calculamos su negatividad, tenemos negatividades del orden de 10^-6 y 10^-7. Creo que se trata de errores numéricos, porque, por ejemplo, el test de traza unidad falla para tolerancias iguales o menores de 10^-6. Hay que subir a tolerancia de 10^-5 para que el test sea exitoso sobre todas las matrices. Así pues, no debe de extrañarme que tanto la traza unidad como la nulidad de la negatividad sean ambas exactas hasta orden O(10^-5).

### 3.3.2.

Nota que la transposición parcial preserva la hermiticidad del sistema. Pues la transposicion en una de las partes preserva la hermiticidad (y ya sabemos que tambien preserva la positividad), luego el producto tensorial sigue siendo hermítico. Ahora bien, ya sabemos que si la representación matricial con la que estamos trabajando no es realmente separable, entonces la operación que realizamos de conmutar elementos no se corresponde realmente con una transposición parcial (porque la transposición parcial no está bien definida en ese caso). Bueno pues si lo compruebas a lo bruto, puedes ver que en ambos casos las operaciones de conmutación de elementos que hacemos en dichas representaciones matriciales también preservan la hermiticidad. Así pues, puedo seguir utilizando np.lingalg.eigh para calcular sus autovalores y, con ellos, la negatividad.
